```python
import numpy as np
from scipy import stats


# Notas de Matemática
notas = [78, 85, 91, 73, 88, 92, 79, 85, 90, 87]

# Calcular a média
media = np.mean(notas)
print(f"Média: {media}")
# Calcular a mediana
mediana = np.median(notas)
print(f"Mediana: {mediana}")
# Calcular a moda
moda = stats.mode(notas, keepdims=True)
print(f"Moda: {moda.mode[0]}, Frequência: {moda.count[0]}")








```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Carregar o dataset
dados = pd.read_csv('notas_30_alunos.csv')

# Calcular medidas de tendência central para cada disciplina
disciplinas = ['matematica', 'ciencias', 'ingles']
resultados = {}

for disciplina in disciplinas:
    media = np.mean(notas)
    mediana = np.median(notas)
    moda = stats.mode(notas, keepdims=True)
    resultados[disciplina] = {
        'Média': media,
        'Mediana': mediana,
        'Moda': moda.mode[0]
    }
# Exibir os resultados
for disciplina, medidas in resultados.items():
    print(f"{disciplina}:")
    for medida, valor in medidas.items():
        print(f"  {medida}: {valor}")

# Visualizar os resultados com gráficos
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
for i, disciplina in enumerate(disciplinas):
    axs[i].hist(dados[disciplina], bins=10, edgecolor='black')
    axs[i].set_title(f'Distribuição de {disciplina}')
    axs[i].set_xlabel('Notas')
    axs[i].set_ylabel('Frequência')


plt.tight_layout()
plt.show()





```
